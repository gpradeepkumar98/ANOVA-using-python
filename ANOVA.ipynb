# **ANOVA**

```python
import pandas as pd
import io
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['ANOVA_2.csv']))
```

```python
df
```

```python
#null hypothesis = All fertilizers are same
#Alternate hypothesis = Atleast one fertilizer is different
```

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(df['Fertilizer_1'], df['Fertilizer_2'], df['Fertilizer_3'])
print(round(fvalue,2), round(pvalue,4))
```

```python
#The p value obtained from ANOVA analysis is significant (p < 0.05), and therefore,
#we conclude that there are significant differences among treatments.
```

# **Tukey test**

```python
# reshape the d dataframe suitable for statsmodels package
df_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['Fertilizer_1', 'Fertilizer_2', 'Fertilizer_3'])
# replace column names
df_melt.columns = ['index', 'treatments', 'value']
```

```python
df_melt
```

```python
pip install bioinfokit
```

```python
# perform multiple pairwise comparison (Tukey's HSD)
from bioinfokit.analys import stat

res = stat()
res.tukey_hsd(df=df_melt, res_var='value', xfac_var='treatments', anova_model='value ~ C(treatments)')
res.tukey_summary
```

```python
#Above results from Tukey’s HSD suggests that except Fertilizer_2 and	Fertilizer_3, all other pairwise comparisons for treatments rejects null hypothesis (p < 0.05)
#and indicates statistical significant differences.
```
